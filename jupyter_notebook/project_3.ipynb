```python
# 컬럼 정의

# Poster_Link - Link of the poster that imdb using (영화 포스터)
# Series_Title = Name of the movie (영화 제목)
# Released_Year - Year at which that movie released (영화 개봉 년도)
# Certificate - Certificate earned by that movie (???)
# Runtime - Total runtime of the movie (러닝 타임)
# Genre - Genre of the movie (영화 장르)
# IMDB_Rating - Rating of the movie at IMDB site (IMDB 사이트의 영화 등급)
# Overview - mini story/ summary (영화 줄거리/요약)
# Meta_score - Score earned by the movie (영화 점수)
# Director - Name of the Director (감독명)
# Star1,Star2,Star3,Star4 - Name of the Stars (배우 이름)
# Noofvotes - Total number of votes (평가 참여자 수)
# Gross - Money earned by that movie (영화 수익 = $)
```

```python
# 장르별로 보기
# 평점순으로 보기 (Meta_score & IMDB_Rating)
# 러닝타임 순으로 보기 (기준 = 분(min))
# 어떤 장르의 비율이 높은지 = 차트
```

```python
# 메타스코어(Meta score)란 ?
# 북미의 영화, TV 프로그램, 게임 및 음악 리뷰 모음 집계 사이트인
# 메타크리틱(Metacritic)에서
# 각종 신문이나 잡지 등의 매체에서 올라온 리뷰들을 통계화하여 집계한 점수
# => 메타크리틱에서 집계한 해당 작품의 평점
```

```python
# IMDB 란 ?
# Internet Movie Database 의 약자
# 미국의 영화 정보 모음 사이트
# 현재 아마존닷컴의 자회사이며, 하위 서비스로 박스오피스 모조가 있음.
# 콜 니덤(Col Needham)이라는 프로그래머가 영화 《죠스》를 본 뒤에
# 영화광이 되어서 인터넷 사업 초창기이던 1990년에 사이트를 개설한 것이 시초.
# 1998년 4월 아마존닷컴이 IMDb를 인수하고,
# 2008년 아마존닷컴이 박스오피스 모조를 인수하면서 IMDb에 합병하였다.
```

```python
genre_percent = []
for genre in Genre_list :
    genre_sum = df['Genre'].str.contains(genre).sum()
    percent = (genre_sum/714) * 100
    genre_percent.append(percent)
print(genre_percent)
```

```python
df_genre = pd.DataFrame()
```

```python
df_genre['Genre'] = Genre_list
```

```python
df_genre['%'] = genre_percent
```

```python
df_genre.sort_values('%', ascending=False, inplace=True)
```

```python
df_genre = df_genre.reset_index().drop('index', axis=1)
```
